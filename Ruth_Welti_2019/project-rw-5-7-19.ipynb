In the cell below, you are asked to name the input and output files.
<br>

**Head group input:** Three possible input files are provided for head group information. 'annotation_input1.csv' is a file with 16 head groups. 'annotation_input1a.csv' is a file with 6 head groups (common phospholipids only). 'annotation_input1b.csv' is a file with 2 head groups (galactolipids). 
Information on the adduct is also included with the head groups.  This is necessary because some lipids form multiple adducts in the mass spectrometer, and the different adducts have different masses. The head group is the part of the molecule that determines which adducts can form.
Head group abbreviations: PA: phosphatidic acid, PC: phosphatidylcholine, PE: phosphatidylethanolamine, PG: phosphatidylglycerol, PI: phosphatidylinositol, PS: phosphatidylserine, MGDG: monogalactosyldiacylglycerol, DGDG: digalactosyldiacylglycerol, TrGDG: trigalactosyldiacylglycerol, TeGDG: tetragalactosyldiacylglycerol.  
<br>

**Fatty acid input:** Three possible input files are provided with fatty acid information. 'annotation_input2.csv' is a file with 100 fatty acids.  It can be used, but it makes a graph that's very hard to read. 'annotation_input2a.csv' is a file with 20 common fatty acids.  'annotation_input2b.csv' is a file with the 3 most common fatty acids in galactolipids. Fatty acid names are total carbons: total carbon:carbon double bonds. Oxidized fatty acids end with "-O" or "-2O" to indicate extra oxygens. The user can create a custom list by choosing the fatty acids of interest (from 'annotation_input2.csv').
<br>

Using smaller input files (like those ending in a and b) makes it easier to see the labels on the results in the scatter plot.

<br>
**Outputs:** Data you will receive will be a csv file with the names of all the lipids formed from combining a head group and two fatty acids and a image file showing a graph of intact ion mass vs fatty acid mass with each annotation plotted and labeled on the graph. Output names are up to you.

```python
hg = input("Type the name of the file with the head group data, ending in .csv")
fa = input("Type the name of the file with the fatty acid data, ending in .csv")
output_csv = input("Type the name of the output csv file, ending in .csv")
output_jpg = input("Type the name of the output image file, ending in .jpg")
```

The cell below creates a dataframe from the input csv file with information about the head groups/adducts. The masses in the this file were calculated with no charge.
Column labels or the data frame can viewed by removing the # on one of the last two lines.

```python
df1 = pd.read_csv(hg)
#list(df1.columns.values)
#df1
```

The cell below creates a dataframe from the input csv file with information about the fatty acids. It can viewed by removing the hashmark on the last line.
Fatty acid masses are the anion masses but they are calculated directly from the formulas and do not account for the extra electron in the anion.
Column labels or the data frame can viewed by removing the # on one of the last two lines.

```python
df2 = pd.read_csv(fa)
#list(df2.columns.values)
#df2
```

The code in the cell below creates a list of lipid names by concatenating the head group/adduct information with the fatty acid information.
Putting the fatty acid in parentheses, separated by a slash, is standard formatting for lipid names.
The output is a list of lipid names, which can be viewed by removing the hashmark on last line.

```python
lipid_names = []
for a in df1['Head_group_adduct']:
    for b in df2['Fatty_acid']:
        for c in df2['Fatty_acid']:
            lipid_names.append(a + "(" + b + "/" + c + ")")
#print(lipid_names)
```

The code in the cell below calculates the mass/charge value for a lipid by adding head group, two fatty acids, and the mass of an electron. The mass of an electron, 0.00054, was added to the total of the intact ion mass and to the anion mass to cover the electron mass, which is part of the whole molecule and ends up in the fatty acyl anion upon fragmentation. Outputs are lists of lipid masses and acyl masses that can be viewed by removing the hashmark on the last two lines. The last fatty acid in the lipid name is the one whose mass is reported.

```python
lipid_mass = []
acyl_mass = []
for i in df1['HG_mass']:
    for j in df2['Acyl_mass']:
        for k in df2['Acyl_mass']:
            lipid_mass.append(i + j + k + 0.00054)
            acyl_mass.append(k + 0.00054)
# print(lipid_mass)
# print(acyl_mass)
```

In the cell below, a dictionary is created from the outputs in the two cells above, and the dictionary is used to create a new dataframe with a list of the lipid names, their (negatively charged) masses and the mass of one of the fatty acyl anions making up the lipid.
The dataframe headers can be viewed by removing the hashmark on line 3.
The name of the output file is the one the user indicated above. If you need to store it somewhere else, you may need to change the path in line 4.

```python
lipid_dict = {'Lipid_names':lipid_names,'Lipid_mass':lipid_mass, 'Acyl_mass': acyl_mass}
df = pd.DataFrame(lipid_dict)
#list(df.columns.values)
df.to_csv(r'/Users/Ruth/Dropbox/coding/semifinal-project-4-19/' + output_csv)
df
```

In the cell below, a scatter plot is made and exported.
You may need to zoom in on the plots to read the labels.  Even then, some lipids have very similar locations in the plot, so the labels may overlap.
I put an example of the plots from the large files (ending in 1 and 2), the mediun files (ending in 1a and 2a), and the small files (ending in 1b and 2b) below the output of the code.  It's also possible to mix and match any head group file (here the files ending in "1") with and any fatty acid file (here the files ending in 2), as long as the input csv files have the same format as the examples.

```python
plt.figure(figsize=(160,120))
ax = df.set_index('Lipid_mass')['Acyl_mass'].plot(style='o')
plt.ylabel("Fragment (fatty acid), m/z", fontsize=80)
plt.xlabel("Intact ion, m/z", fontsize=80)
plt.rcParams['ytick.labelsize'] = 80.0
plt.rcParams['xtick.labelsize'] = 80.0

def label_point(x, y, z, ax):
    a = pd.concat({'Lipid_mass': x, 'Acyl_mass': y, 'Lipid_names': z}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['Lipid_mass'], point['Acyl_mass'], str(point['Lipid_names']), rotation = 315)

label_point(df.Lipid_mass, df.Acyl_mass, df.Lipid_names, ax)

ax.set_ylim([100,400])
ax.set_xlim([300,1600])
plt.savefig(output_jpg) 
plt.show()
```

Below, I added a graph in Bokeh in which one can mouse over the points to see the annotation infomation.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.plotting import output_file
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool
output_notebook()
```

```python
source = ColumnDataSource(df)

hover = HoverTool(tooltips=[
    ("Lipid mass", "@Lipid_mass"),
    ("Acyl mass", "@Acyl_mass"),
    ("Lipid names", '@Lipid_names'),
])

p = figure()
p = figure()
p.plot_width = 600
p.plot_height = 600
p.tools = [hover]
p.xaxis.axis_label = "Lipid mass"
p.yaxis.axis_label = "Acyl mass"
p.xgrid.minor_grid_line_color = 'navy'
p.xgrid.minor_grid_line_alpha = 0.1
p.ygrid.minor_grid_line_color = 'navy'
p.ygrid.minor_grid_line_alpha = 0.1


p.scatter('Lipid_mass', 'Acyl_mass', source = source)
show(p)
```
