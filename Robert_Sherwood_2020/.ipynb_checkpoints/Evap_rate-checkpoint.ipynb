# Prepare Data for Hydrus-1D

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from scipy import stats
```

```python
#Reading raw evaporation data.
df = pd.read_csv('..\AGRON 935\Evaporation.csv', sep = '\t')

#Defining the dataframe of interest to be used for calculaitng the evaporation rate.
start = 22
end = 108
df = df[start:end]

#Giving a meaningful names to a column of data.
df['Mass'] = df['Unnamed: 3']
df['Time AM/PM'] = df['Unnamed: 1']

#Converting time from [[YYYY/MM/DD], [HH:MM AM/PM]] to [sec] for calculations and plotting.
df['Datetime'] = pd.to_datetime(df['Date'] + " " + df['Time AM/PM'])          #from [[MM/DD/YYYY], [HH:MM, AM/PM]] to [YYYY/MM/DD, HH:MM:SS].
df['Time_diff']= df.Datetime-df.Datetime[start]                               #from [YYYY/MM/DD, HH:MM:SS] to [days, HH:MM:SS].
df['Delta_sec']= df['Time_diff'].dt.days * 86400 + df['Time_diff'].dt.seconds #from [days, HH:MM:SS] to [sec].

#Fit a regression line to the measured evaportaion data. The slope of the line is the evaporation rate in gr./sec.
fit_info = stats.linregress(df.Delta_sec[start:len(df.Delta_sec)], (df.Mass[start:len(df.Mass)]))
#Converting the evaportation rate (i.e., slope) from gr./sec to gr./day
Evap_rate = fit_info.slope * 86400 #[gr./day]

#Create strings containing the regression coeffecient and evaportion rate. These strings are used
#to report the values inside of the figure at the end of the code.
R_2 = 'R^2 = ' + str(round((fit_info.rvalue**2),2))
Evap = 'Evaporation rate = ' + str(round(Evap_rate, 4)) + ' gr./day'

#Print the regression coefficient (R^2), and the calculated evaportaion rate (Evap) in gr./day.
print(R_2), print(Evap), print(' ')

#Calculation of the predicted evaporation of water verses time. Equation form: y = m*x + b.
df['Mass_predict'] = fit_info.slope * df.Delta_sec + fit_info.intercept #[grams]

#Pot figure of Evaportaion Data with fitted regression line.
plt.plot(df.Delta_sec, (df.Mass_predict), color = 'k', label = 'Evap. Pred.')
plt.scatter(df.Delta_sec, (df.Mass), alpha = 0.5, label='Evap. Obs.')
plt.ylabel('Mass [gr.]', fontsize = 16)
plt.xlabel('Time [sec]', fontsize = 16)
plt.xlim(df.Delta_sec[start], df['Delta_sec'].max())
plt.text(250000, .15, 'Rate of Evaporation', fontsize = 20, horizontalalignment ='center', verticalalignment ='center')
plt.text(7000, -.40, Evap , fontsize = 10, horizontalalignment ='left', verticalalignment ='center')
plt.text(7000, -.44, R_2 , fontsize = 10, horizontalalignment ='left', verticalalignment ='center')
plt.legend()
plt.tight_layout()
plt.show()
```
